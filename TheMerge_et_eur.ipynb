```python
import pandas as pd
import numpy as np
import datetime as dt
```

```python
df_open_merged = pd.read_csv('/content/crypto_et_open_merged.csv')
df_close_merged = pd.read_csv('/content/crypto_et_closed_merged.csv')
```

```python
df_open_merged['Date'] = pd.to_datetime(df_open_merged['Date'],dayfirst=True)
df_open_merged['Open Date'] = pd.to_datetime(df_open_merged['Open Date'],dayfirst=True)
df_close_merged['Date'] = pd.to_datetime(df_close_merged['Date'],dayfirst=True)
df_close_merged['Close Date'] = pd.to_datetime(df_close_merged['Close Date'],dayfirst=True)
```

```python
print(df_close_merged.columns)
print(df_close_merged.shape)
print(df_open_merged.columns)
print(df_open_merged.shape)
print(df_close_merged.dtypes)
print(df_open_merged.dtypes)
```

```python
print(df_close_merged['Position ID'].nunique())
print(df_open_merged['Position ID'].nunique())
```

```python
df_open_merged['Amount_Eur'] = df_open_merged['Amount']/df_open_merged['Low']
df_open_merged['Open Rate_Eur'] = df_open_merged['Open Rate']/df_open_merged['Low']
df_close_merged['Close Rate_Eur'] = df_close_merged['Close Rate']/df_close_merged['High']
df_close_merged['Amount_close_Eur'] = (df_close_merged['Close Rate_Eur']*df_close_merged['Units']) - (df_close_merged['Spread']/df_close_merged['Low'])
```
